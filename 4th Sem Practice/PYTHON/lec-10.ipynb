```python
# Write a program which will print sum of all the keys whose values are the factor of 12

dt = {1:2, 2:3, 3:4, 6:12, 17:18, 18:12}
total = 0
for keys,values in dt.items():
    if 12 % values == 0:
        total += keys
print(total)        
```
