```python
# Q1. Write a Python program to check if a given string is a valid email address.
# Ex: example@domain.com
# Hint: Check ‘@’ comes earlier than ‘dot’

x=input("Enter the email address: ")
if '@' in x and '.' in x and x.index('@') < x.index('.'):
    print("Valid email address")
else:
    print("Invalid email address")
```
