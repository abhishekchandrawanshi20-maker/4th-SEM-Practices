```python
# Q7: Write a Python program to print all the student names in the given dictionary.

dt={'Alice': 85, 'Bob': 90, 'Charlie': 80}
for k in dt.keys():
    print(k)
```

```python
# Q8: Write a Python program to print all the grades in the given dictionary.

dt={'Alice': 85, 'Bob': 90, 'Charlie': 80}
for v in dt.values():
    print(v)
```

```python
# Q9: Write a Python program to print all key-value pairs in the given dictionary in the format “Student: Grade”.


dt={'Alice': 85, 'Bob': 90, 'Charlie': 80}
for k,v in dt.items():
    print(k,v)
```
