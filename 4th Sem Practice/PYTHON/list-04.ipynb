```python
# Write a program which will take list as a input from and increase all the even no. by 5 and decrease all the odd no. by 5



lst=eval(input("Enter a list: "))
print(lst)
for i in range(len(lst)):
    if lst[i]%2 == 0:
        lst[i] += 5
    else:
        lst[i] -= 5
print(lst)
```
