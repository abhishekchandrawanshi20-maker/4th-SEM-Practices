```python
#  odd index pe list -> odd values => +1
#  even index pe list -> even values => -1
lst=[[3,2,1,8], [50,3,7,9], [2,3,7,8]]
ans=[]
for i in range(len(lst)):
    temp=lst[i].copy()
    if i%2==0:
        for j in range(len(temp)):
            if temp[j]%2==0:
                temp[j]-=1
        ans.append(temp)
    else:
        for j in range(len(temp)):
            if temp[j]%2!=0:
                temp[j]+=1
        ans.append(temp)
print(lst)
print(ans)                                    
```
